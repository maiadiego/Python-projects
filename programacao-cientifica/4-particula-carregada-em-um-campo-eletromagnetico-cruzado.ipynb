Vamos observar a trajetória de uma partícula elétrica carregada em um campo elétrico uniforme cruzado com um campo magnético em um ângulo de 90 graus.

```python
# bibliotecas utilizadas
import numpy as np
import matplotlib.pyplot as plt 
from scipy.integrate import solve_ivp
from IPython.display import display, Math, Latex
```

```python
# função que descreve o movimento da partícula a partir de equações diferenciais dadas
def motion_particle(t, y, E, B, q):    
    dy_dt = y[1] # vy 
    dz_dt = y[3] # vz
    
    dvy_dt = q * B * y[3]
    dvz_dt = q * (E - (B*y[1]))
    
    return [dy_dt, dvy_dt, dz_dt, dvz_dt]
```

```python
# parâmetros
q = 1 # carga da partícula elétrica
m = 1 # massa
E = 1 # const
B = 1 # const campo magnético
```

```python
# set tempos
t_span = np.array([0, 50]) # intervalo de integração
times = np.linspace(t_span[0], t_span[1], 1000) # passagem de tempo
```

```python
# set condições iniciais
y0 = 0
vy0 = 0
z0 = 0
vz0 = 0 
Y0 = [y0, vy0, z0, vz0]
```

```python
# encontrar a solução resolvendo as equações diferenciais
sol = solve_ivp(motion_particle, t_span, y0=Y0, t_eval=times, args=(E, B, q))
```

```python
# plotar o gráfico 
plt.figure(figsize=(10,7))
plt.plot(sol.y[0], sol.y[2], '-b')
plt.style.use('ggplot')
plt.title("Trajetória de uma partícula em um campo eletromagnético cruzado", fontsize=16)
plt.xlabel('y - Intervalo de tempo', fontsize=14)
plt.ylabel('z - Posição da partícula', fontsize=14)
plt.tight_layout()
plt.show()
```

O período do movimento da partícula nesse campo eletromagético cruzado é dado por T=2pi*m/q*B. Portanto, igual a 2pi, aproximadamente 6 segundos.

```python
# plotando o gráfico do tempo
R = np.sqrt ((sol.y[0] - sol.t)**2 + (sol.y[2] - 1)**2)
plt.figure(figsize=(10,7))
plt.axis ([0, 50, 0.5, 1.5])
plt.plot (sol.t, R)
plt.style.use('ggplot')
plt.title("R(t)", fontsize=16)
plt.tight_layout()
plt.show ()
```
